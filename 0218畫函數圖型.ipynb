#畫函數圖形
我想利用多個玫瑰線疊成蓮花，查詢到玫瑰圖的方法，原本只是簡單粉色紫色不太像，又去查了漸層的方法。
過程中也有試過一般常見的那種圓形花瓣花朵，但最後這版最像所以保留。
再加上方向旋轉製造交錯的感覺。


```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

# 利用多個玫瑰線疊成蓮花，查詢:https://zh.wikipedia.org/zh-tw/%E7%8E%AB%E7%91%B0%E7%BA%BF
# 設定參數
a = 1  # 蓮花大小
k = 10  # 花瓣數量，2k 個花瓣
layers = 5  # 層數
rotation_angle = np.pi /layers  # 每次畫圖時的旋轉角度（弧度）

# 製作漸變淺粉到深粉
colors = ["#ffebf5", "#e40072"]  # 製作淺粉到深粉漸層，方法查詢:https://icolorpalette.com/color/ff69b4
cmap = LinearSegmentedColormap.from_list("pink_gradient", colors, N=layers)

# 定義角度範圍
theta = np.linspace(0, 2 * np.pi, 1000) # 2pi=360度

# 繪製蓮花圖
plt.figure(figsize=(6, 6))
ax = plt.subplot(projection='polar')  # 使用極坐標系
# 使用迴圈繪製每一層
for i in range(layers):
    # 花瓣的長度 r=cos(k* theta)的話，花會是正的!
    r = (a + i * 0.2) * np.sin(k * theta)

    # 花瓣交錯，增加角度偏移
    offset = i * rotation_angle  # 每一層增加旋轉角度
    r_rotated = r * np.cos(offset)  # 用cos進行x軸旋轉

    # 根據不同的層次給每一層不同顏色
    color = cmap(i / layers)  # 顏色切成好幾份
    ax.plot(theta, r_rotated, color=color, alpha=0.8)  # 設定顏色和透明度


# 設定圖形
ax.set_title("Lotus Flower")
ax.grid(True)

# 顯示圖形
plt.show()
```
