#畫函數圖形



```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 400) #從-5到5，總共400個點，讓曲線更順滑

y = np.abs(x)

plt.plot(x, y )#開始畫圖

#幫x、y軸加上標籤
plt.xlabel('x')
plt.ylabel('|x|')
plt.title('y = |x|')
plt.grid()
```
