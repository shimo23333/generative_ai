```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

# :https://zh.wikipedia.org/zh-tw/%E7%8E%AB%E7%91%B0%E7%BA%BF
# 
a = 1  # 
k = 10  # 2k 
layers = 5  # 
rotation_angle = np.pi /layers  # 

# 
colors = ["#ffebf5", "#e40072"]  # :https://icolorpalette.com/color/ff69b4
cmap = LinearSegmentedColormap.from_list("pink_gradient", colors, N=layers)

# 
theta = np.linspace(0, 2 * np.pi, 1000) # 2pi=360

# 
plt.figure(figsize=(6, 6))
ax = plt.subplot(projection='polar')  # 
# 
for i in range(layers):
    #  r=cos(k* theta)!
    r = (a + i * 0.2) * np.sin(k * theta)

    # 
    offset = i * rotation_angle  # 
    r_rotated = r * np.cos(offset)  # cosx

    # 
    color = cmap(i / layers)  # 
    ax.plot(theta, r_rotated, color=color, alpha=0.8)  # 


# 
ax.set_title("Lotus Flower")
ax.grid(True)

# 
plt.show()
```
